```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('park.jpg',0)
  

equ = cv2.equalizeHist(img)

row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].hist(img.ravel(), 256, [0, 256])
axs[0].set_title('Histogram')
 
axs[1].imshow(cv2.cvtColor(equ, cv2.COLOR_BGR2RGB))
axs[1].set_title('Histogram Equalization')

plt.show()
```
