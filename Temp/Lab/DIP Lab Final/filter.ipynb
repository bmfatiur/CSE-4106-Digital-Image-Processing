```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('park.jpg')
```

```python
#average blur
average_img=cv2.blur(img,(3,3))
#gaussian blur
gauss_img=cv2.GaussianBlur(img,(7,7),0)
#median blur
median_img=cv2.medianBlur(img,3)


row, col = 2, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()

axs[0][0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0][0].set_title('input image')

axs[0][1].imshow(cv2.cvtColor(average_img, cv2.COLOR_BGR2RGB))
axs[0][1].set_title('average blur')
 
axs[1][0].imshow(cv2.cvtColor(gauss_img, cv2.COLOR_BGR2RGB))
axs[1][0].set_title('gaussian blur')

axs[1][1].imshow(cv2.cvtColor(median_img, cv2.COLOR_BGR2RGB))
axs[1][1].set_title('median blur')

plt.show()
```
